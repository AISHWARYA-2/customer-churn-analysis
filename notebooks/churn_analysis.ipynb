```python
import pandas as pd

df = pd.read_csv(
    r"C:\DataAnalysis\customer-churn-analysis\data\WA_Fn-UseC_-Telco-Customer-Churn.csv"
)

print(df.head())
print(df.info())
```

```python
# Missing values
print(df.isnull().sum())

# Convert TotalCharges
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)

df.dropna(inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Churn", data=df)
plt.show()
```

```python
sns.countplot(
    x="gender",
    hue="Churn",
    data=df
)
plt.show()
```

```python
sns.countplot(
    x="Contract",
    hue="Churn",
    data=df
)
plt.xticks(rotation=15)
plt.show()
```

```python
sns.boxplot(
    x="Churn",
    y="MonthlyCharges",
    data=df
)
plt.show()
```
